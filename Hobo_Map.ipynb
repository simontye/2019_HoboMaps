```python
import pandas as pd
import folium
import numpy as np
from folium import plugins
from branca.element import Template, MacroElement
```

```python
hobo_g=pd.read_csv("/Users/simontye/Documents/Research/Projects/NYC_Homeless_Shelters/Hobo_Government.csv")
hobo_n=pd.read_csv("/Users/simontye/Documents/Research/Projects/NYC_Homeless_Shelters/Hobo_NonProfit.csv")
hobo_p=pd.read_csv("/Users/simontye/Documents/Research/Projects/NYC_Homeless_Shelters/Hobo_Private.csv")
hobo_r=pd.read_csv("/Users/simontye/Documents/Research/Projects/NYC_Homeless_Shelters/Hobo_Religious.csv")

hobo_g=hobo_g.fillna('NA')
hobo_n=hobo_n.fillna('NA')
hobo_p=hobo_p.fillna('NA')
hobo_r=hobo_r.fillna('NA')

hobo_g.head()
```

```python
hobo_map=folium.Map(location=[40.735, -73.98],
    zoom_start=13,
    max_zoom=17,
    detect_retina=True,
    tiles='http://maps.nypl.org/warper/maps/tile/16200/{z}/{x}/{y}.png',
    attr='Manhattan, 1958. New York Public Library.')

for index, row in hobo_g.iterrows():
    popup_text="{}<br> Address: {}<br> Director: {}<br> Capacity: {}<br> Priority: {}<br> Aid: {}"
    popup_text=popup_text.format(row["Name"],row["Address"],row["Director"],row["Capacity"],row["Priority"],row["Aid"])
    folium.Marker(icon=folium.Icon(icon='university',color="black",prefix='fa'),location=(row["Latitude"], row["Longitude"]),popup=popup_text).add_to(hobo_map)

for index, row in hobo_n.iterrows():
    popup_text="{}<br> Address: {}<br> Director: {}<br> Capacity: {}<br> Priority: {}<br> Aid: {}"
    popup_text=popup_text.format(row["Name"],row["Address"],row["Director"],row["Capacity"],row["Priority"],row["Aid"])
    folium.Marker(icon=folium.Icon(icon='building',color="blue",prefix='fa'),location=(row["Latitude"], row["Longitude"]),popup=popup_text).add_to(hobo_map)

for index, row in hobo_p.iterrows():
    popup_text="{}<br> Address: {}<br> Director: {}<br> Capacity: {}<br> Priority: {}<br> Aid: {}"
    popup_text=popup_text.format(row["Name"],row["Address"],row["Director"],row["Capacity"],row["Priority"],row["Aid"])
    folium.Marker(icon=folium.Icon(icon='building',color="orange",prefix='fa'),location=(row["Latitude"], row["Longitude"]),popup=popup_text).add_to(hobo_map)

for index, row in hobo_r.iterrows():
    popup_text="{}<br> Address: {}<br> Director: {}<br> Capacity: {}<br> Priority: {}<br> Aid: {}"
    popup_text=popup_text.format(row["Name"],row["Address"],row["Director"],row["Capacity"],row["Priority"],row["Aid"])
    folium.Marker(icon=folium.Icon(icon='users',color="green",prefix='fa'),location=(row["Latitude"], row["Longitude"]),popup=popup_text).add_to(hobo_map)
```
